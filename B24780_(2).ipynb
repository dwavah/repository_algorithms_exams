Question 2(Plotting Algorithms)

```python

#import necessary libraries
import time
import random
import matplotlib.pyplot as plt

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

def selection_sort(arr):
    for i in range(len(arr)):
        min_index = i
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]

def measure_time(sorting_function, input_size):
    lst = [random.randint(1, 100) for _ in range(input_size)]
    start_time = time.time()
    sorting_function(lst)
    end_time = time.time()
    return end_time - start_time

input_sizes = [1, 10, 100, 600, 2000]  # You can adjust the input sizes as needed
insertion_times = []
selection_times = []

for size in input_sizes:
    insertion_time = measure_time(insertion_sort, size)
    insertion_times.append(insertion_time)

    selection_time = measure_time(selection_sort, size)
    selection_times.append(selection_time)

# Plotting the results
plt.plot(input_sizes, insertion_times, label='Insertion Sort')
plt.plot(input_sizes, selection_times, label='Selection Sort')
plt.xlabel('Input Size')
plt.ylabel('Time (seconds)')
plt.title('Performance')
plt.legend()
plt.show()
```
